```python
import sympy as sp
import numpy as np
sp.init_printing()

phi = sp.symbols("phi")
tau_x = np.array([[0,1],[1,0]])
tau_z = np.array([[1,0],[0,-1]])
sigma_0 = np.array([[1,0],[0,1]])
A = sp.Matrix(-1j*sp.cos(phi/2)*np.kron(tau_x,sigma_0)-sp.sin(phi/2)*np.kron(tau_z, sigma_0))
A
```

```python
A.eigenvals()
```
